```python
from utils.filemanagement import load_from_file,write_topics_file

folder_path_data = r"preprocessed_data"
file_name_data ="np_mini_stopwords.csv"

df = load_from_file(folder_path_data,file_name_data)


data = df.text.values.tolist()
```

```python
from nltk.tokenize import word_tokenize
r = []
new_data = []
for item in data:
    new_data.append(str(item))
for item in new_data:
    r.append(word_tokenize(item))

```

```python
print(r[0])
```

```python

# Calculate and save, or load, RAW data word frequencies to/from file
folder_path_frequencies = r"preprocessed_data/stats"

file_name_frequencies="np_mini_stopwords_freq"


word_frequencies = load_word_frequencies(folder_path_frequencies, file_name_frequencies)
if not word_frequencies:
    word_frequencies = word_frequency({}, r)
    write_word_frequencies_file(folder_path_frequencies, file_name_frequencies, word_frequencies)

example_frequency = random.choice(list(word_frequencies.keys()))
print(f"{example_frequency}: {word_frequencies[example_frequency]}")
```

```python
from utils.filemanagement import load_zipped_word_co_frequencies

# Calculate and save, or load, word co-frequencies to/from file

file_name_co_frequencies = f"np_mini_stopwords_co_freq"

#file_path_co_frequencies = os.path.join(ROOT_PATH, folder_path_frequencies, str(file_name_co_frequencies + ".pkl.zip"))

word_co_frequencies = load_zipped_word_co_frequencies(folder_path_frequencies,file_name_co_frequencies)
if not word_co_frequencies:
    word_co_freq_counter = Counter()
    docs = r
    word_co_frequencies = word_co_frequency(word_co_freq_counter,docs)
    write_word_co_frequencies_zip_file(folder_path_frequencies,file_name_co_frequencies,word_co_frequencies)
    #write_word_co_frequencies_file(folder_path_frequencies,file_name_co_frequencies,word_co_frequencies)
```
